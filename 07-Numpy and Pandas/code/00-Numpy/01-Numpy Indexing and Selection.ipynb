## 1) Indexing
The simplest way to pick one or some elements of an array looks very similar to python lists:

### 1.1) Indexing a 1D array (Vectors)

```python
#Get a value at an index
arr[8]
```

```python
#Get values in a range
arr[1:5]
```

```python
#Get values in a range
arr[0:5]
```

```python
arr[:5]
```

```python
arr[3:]
```

```python
arr[3:7]
```

```python
arr[3:7:2]
```

**if you want to make operation on a copy of the array and don't change the base array just use .copy()**


```python
arr = np.arange(30,40)
arr
```

```python
#Setting a value with index range (Broadcasting)
arr[0:5] = 100

#Show
arr
```

```python
# Reset array, we'll see why I had to reset in  a moment
arr = np.arange(30,40)

#Show
arr
```

```python
#Important notes on Slices
slice_of_arr = arr[0:6]

#Show slice
slice_of_arr
```

```python
#Change Slice
slice_of_arr[:] = 99

#Show Slice again
slice_of_arr
```

Now note the changes also occur in our original array!

```python
arr
```

Data is not copied, it's a view of the original array! This avoids memory problems!

```python
myarr = np.arange(30,40)
myarr
```

```python
arr = myarr.copy()
arr
```

```python
slice_of_arr = arr[0:6]

#Show slice
slice_of_arr
```

```python
slice_of_arr[:] = 99

#Show Slice again
slice_of_arr
```

```python
arr
```

```python
myarr
```

### 1.2) Indexing a 2D array (matrices)

The general format is **arr_2d[row][col]** or **arr_2d[row,col]**. I recommend usually using the comma notation for clarity.

```python
arr_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))

#Show
arr_2d
```

```python
#Indexing row
arr_2d[1]
```

```python
# Format is arr_2d[row][col] or arr_2d[row,col]

# Getting individual element value
arr_2d[1][0]
```

```python
# Getting individual element value
arr_2d[1,0]
```

```python
arr_2d
```

```python
# 2D array slicing

#Shape (2,2) from top right corner
arr_2d[:2,1:]
```

```python
#Shape bottom row
arr_2d[2]
```

```python
#Shape bottom row
arr_2d[2,:]
```

```python
arr_2d
```

```python
arr_2d[::-1,::]
```

```python
arr_2d[:,::-1]
```

```python
arr_2d[::-1,::-1]
```

### 1.3) Fancy Indexing

Fancy indexing allows you to select entire rows or columns out of order,to show this, let's quickly build out a numpy array:

```python
#Set up matrix
arr2d = np.zeros((10,10))
```

```python
arr2d
```

```python
for i in range(10):
    arr2d[i] = i + 2
    
arr2d
```

Fancy indexing allows the following

```python
arr2d[[1,4,6,8]]
```

```python
#Allows in any order
arr2d[[6,4,2,7]]
```

```python
arr2d[[1,4,6,8], 2:5]
```

## 2) Selection

Let's briefly go over how to use brackets for selection based off of comparison operators.

```python
arr = np.arange(1,10)
arr
```

```python
arr > 4
```

```python
bool_arr = arr > 4
```

```python
bool_arr
```

```python
arr[bool_arr]
```

```python
arr
```

```python
arr[arr >= 5]
```

```python
x = 3
arr[arr > x]
```

```python
arr = np.arange(1,10)
arr
```

```python
arr[arr % 2 == 0]
```
