## Lets start Playing with SQL

**We already scrapped data in first notebook and created the database with the above structure and populate it with the scrapped data**

```python
import sqlite3

conn = sqlite3.connect('books.db')
cur = conn.cursor()
```

```python
def get_rows(sql, data):
    global conn
    global cur
    cur.execute(sql, data)
    rows = cur.fetchall()
    for ID, title, price, rate, cat_id in rows:
        print(f"[{ID}] ----- {title} ----- {price} £----- {rate} ----- Cat_id: {cat_id}")
```

**01) Write a sql query to get books that has rate >=3 and has 'Mr' in its name.**

```python
get_rows("SELECT * FROM books WHERE rate >= ? AND title LIKE ?", [3, '%Mr%'])
```

**02) Write a sql query to get books that has rate >=3 or price > 20.**

```python
cur.execute("SELECT books.title,books.rate,categories.name FROM books INNER JOIN categories WHERE books.category_id = categories.id", [])
rows = cur.fetchall()
print(rows)
```

**15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.**

```python
cur.execute("SELECT COUNT(id), rate FROM books WHERE price BETWEEN ? AND ? GROUP BY rate", [20, 30])
rows = cur.fetchall()
print(rows)
```

**16) Write a sql query to calculate how many books each category has having count > 10.**

```python
cur.execute("SELECT COUNT(id) AS CNT, category_id FROM books GROUP BY category_id HAVING CNT > ?", [10])
rows = cur.fetchall()
print(rows)
```
