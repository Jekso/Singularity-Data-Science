### 2) from, import

```python
from random import shuffle
```

```python
# This shuffles the list "in-place" meaning it won't return
# anything, instead it will effect the list passed

mylist = [40, 10, 100, 30, 20]
shuffle(mylist)
```

```python
mylist
```

```python
shuffle(mylist)
```

```python
mylist
```

**you can import many functions from a module using `,`**

```python
from random import shuffle, randint
```

```python
randint(10, 20)
```

**or you can import all function from module**

```python
from random import *
```

```python
randint(10, 50)
```

```python
my_list = [10, 50, 80]
shuffle(my_list)
my_list
```
