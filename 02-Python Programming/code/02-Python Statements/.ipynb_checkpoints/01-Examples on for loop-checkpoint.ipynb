# Examples on for loop

```python
for i in range(1, 101):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

```python
my_list = [10, 30, 35, 12, 100, -4, -20, 0]
for i in my_list:
    if i == 0:
        print(f"{i} = zero, not even not odd")
    elif i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

```python
my_list = list(range(1, 5))

sum_of_numbers = 0

for i in my_list:
    sum_of_numbers = sum_of_numbers + i
print(sum_of_numbers)

    
    
```
