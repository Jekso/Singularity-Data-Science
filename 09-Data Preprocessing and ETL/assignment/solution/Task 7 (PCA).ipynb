1. Using Salaries.csv dataset to generate PCA components that represents 90% of variance of data, and figure out how many components are, and get each component variance value.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':(10, 10)}, font_scale=1.2)
```

```python
df = pd.read_csv('../../code/dastasets/Salaries.csv')
df
```

```python
df = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']]
df
```

```python
df = df.dropna()
df
```

```python
X = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']]
y = df['TotalPayBenefits']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.9)

X = pca.fit_transform(X)
```

```python
pca.n_components_
```

```python
df = pd.DataFrame(data=X)
df
```
