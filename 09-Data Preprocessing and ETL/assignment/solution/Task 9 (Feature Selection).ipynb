```python
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif
from sklearn.feature_selection import SelectKBest
```

```python
X = df.drop(['not.fully.paid', 'purpose'], axis=1)
y = df['not.fully.paid']
```

```python
X
```

```python
selector = SelectKBest(k=10, score_func=f_classif)
selector.fit(X, y)
```

```python
pd.DataFrame(selector.scores_, index=X.columns)
```

```python
best_features = X.columns[selector.get_support(indices=True)]
best_features
```
