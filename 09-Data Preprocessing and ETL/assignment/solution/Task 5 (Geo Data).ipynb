**1) Using visitor-interests.csv dataset, extract latitude, longitude from 'IP' column and use it to create new 'Dist_From_Egypt' column calculating haversine_distance with Egypt Lat/Lonng, you can get this location from google.**

```python
df = pd.read_csv('../../code/dastasets/visitor-interests.csv')
df = df.head(50)
df
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo

def extract_lat_long_from_ip(ip):
    try:
        response = ip2geo.get(ip, api_key='free')
        return pd.Series((response.latitude, response.longitude))
    except:
        return pd.Series((np.nan, np.nan))
    
    
df[['Latitude', 'Longitude']] = df['IP'].apply(extract_lat_long_from_ip)

df
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
from geopy.distance import great_circle
```

```python
def get_dist(x):
    egypt_lat = 30.033333
    egypt_long = 31.233334
    egypt_loc = (egypt_lat, egypt_long)
    row_loc = (x['Latitude'], x['Longitude'])
    return great_circle(row_loc, egypt_loc).kilometers

df['Dist_From_Egypt'] = df.apply(get_dist, axis=1)
df.head()
```
